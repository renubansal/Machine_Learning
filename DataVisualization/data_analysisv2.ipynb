```python
data1 = df.loc[df['y'] == 'yes']
data2 = df.loc[df['y'] == 'no']

x = data1['marital'].value_counts()
y = data2['marital'].value_counts()

print(x.values,y.values)
print(x.index)
```

```python
import numpy as np
n_group = 3
fig,ax = plt.subplots()
index = np.arange(n_group)
bar_width = 0.3

bar1 = plt.bar(index,x.values,bar_width,color='r',label="yes")
bar2 = plt.bar(index + bar_width,y.values,bar_width,color='b',label="no")

# set title and labels 
plt.title('Marital Status Output-wise') 
plt.xlabel('Marital Status')
plt.legend()
plt.xticks(index+bar_width,x.index,rotation='vertical')
#plt.ylabel('Frequency')
```

## IRIS dataset

```python
iris = pd.read_csv("iris.data",names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
iris.describe()

import matplotlib.pyplot as plt

plt.scatter(iris.sepal_length,iris.sepal_width)
```

```python
# create a figure and axis
fig, ax = plt.subplots()

# scatter the sepal_length against the sepal_width
ax.scatter(iris['sepal_length'], iris['sepal_width'])
ax.scatter(iris['sepal_length'], iris['petal_width'])


# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.legend()
```

```python
colors = {'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}

# create a figure and axis
fig, ax = plt.subplots()

size = len(iris.index)
print(size)
# plot each data-point
for i in range(size):
    ax.scatter(iris['sepal_length'][i], iris['sepal_width'][i], color=colors[iris['class'][i]])
# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
ax.legend(['Iris-setosa','Iris-versicolor','Iris-virginica'])
```

```python
iris.boxplot(column=['sepal_length', 'sepal_width','petal_length','petal_width'], rot=90, fontsize=15)
```

```python
iris['class'].value_counts().plot.pie()
```

```python
plt.figure(figsize=(5,5))
data = iris['class'].values.tolist()
plt.hist(data, facecolor='blue', alpha=0.5)
plt.ylabel('count')
plt.show()
```

```python
columns = iris.columns.drop(['class'])

x_data = range(0, iris.shape[0])
print(x_data)

# create figure and axis
fig, ax = plt.subplots()

# plot each column
for column in columns:
    ax.plot(x_data, iris[column])
# set title and legend
ax.set_title('Iris Dataset')
ax.legend()
```
