## 1. Adding nodes, path graph

```python
G.add_node(1) # add one node
nx.draw(G)
```

## 4. Lets work with an example edgelist

```python
df = pd.read_csv('1999-1-wrsn.tsv', sep = '\t',names=['n1','n2','wgt'])
df.head(n=5)
```

#### Loading this dataframe into a networkx graph

```python
G = nx.from_pandas_edgelist(df,'n1','n2', edge_attr = 'wgt', create_using = nx.MultiDiGraph)
#create simple graph - undirected, no self loops (using Graph instead of MultiDiGraph)
H = nx.from_pandas_edgelist(df,'n1','n2', edge_attr = 'wgt', create_using = nx.Graph)
```

#### Different layouts

```python
# circular layout
nx.draw(G, pos=nx.circular_layout(G), node_color = 'black', node_size = 10, arrow_size = 5, edge_color = "blue", width = 1)
#nx.draw(G, pos=nx.circular_layout(G))
```

```python
# spring layout
nx.draw(G, pos=nx.spring_layout(G), node_color = 'black', node_size = 5, arrow_size = 2, edge_color = "blue", width = 1)
```

```python
# spectral layout
nx.draw(G, pos=nx.spectral_layout(G), node_color = 'black', node_size = 5, arrow_size = 2, edge_color = "blue", width = 1)
```

```python
# random layout
nx.draw(G, pos=nx.random_layout(G), node_color = 'black', node_size = 5, arrow_size = 2, edge_color = "blue", width = 1)
```

### 5.4 Betweenness Centrality 
Betweenness centrality is a measure of the influence of a vertex over the flow of information between every pair of vertices under the assumption that information primarily flows over the shortest paths between them.

The betweenness centrality for each vertex is the number of these shortest paths that pass through the vertex.

A node with higher betweenness centrality would have more control over the network, because more information will pass through that node.

```python
#nx.betweenness_centrality(G)
nx.betweenness_centrality(G)[777] # for specific node
```
